```python
import pandas as pd

df=pd.read_csv('../Data/merged_df_exploration.csv')

print("Shape of the dataset is:", df.shape)
print("\nColumn types and non-null counts:")
print(df.info())

print("\nFirst of few rows:")
display(df.head())
```

```python
missing_counts=df.isnull().sum()
missing_percent=(missing_counts / len(df))*100
missing_df=pd.DataFrame({
    "Missing Count": missing_counts,
    "Missing %": missing_percent
})
print(missing_df[missing_df["Missing Count"] > 0])
```

```python
df = df.dropna().reset_index(drop=True)
print(f"Remaining rows after dropping missing values: {df.shape[0]}")
```

```python
df['avg_mass'] = (df['mass_min'] + df['mass_max']) / 2

df['power_to_weight'] = df['power_hp'] / df['avg_mass']

import numpy as np
df['log_co2'] = np.log1p(df['co2'])  

df.info()
```
