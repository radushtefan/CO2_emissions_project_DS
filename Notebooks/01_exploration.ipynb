```python
float_columns = [
    "cons_urban", "cons_extra", "cons_mixed", 
    "co2", "power_hp", "mass_min", "mass_max", "power_admin"
]

categorical_columns = [
    "brand", "model", "gearbox", "fuel", "hybrid"
]
```

```python
def clean_numerical_columns(df, float_cols):
    for col in float_cols:
        if col in df.columns:
            df[col] = df[col].astype(str).str.replace(',', '.', regex=False)
            df[col] = pd.to_numeric(df[col], errors='coerce')
    return df

def clean_categorical_columns(df, cat_cols):
    for col in cat_cols:
        if col in df.columns:
            df[col] = df[col].astype(str).str.strip().str.upper()
    return df

def clean_brand_column(df):
    if "brand" in df.columns:
        df["brand"] = (
            df["brand"]
            .astype(str)
            .str.strip()
            .str.lower()
            .str.replace("-", " ")
            .str.replace("_", " ")
            .str.replace(r"\s+", " ", regex=True)
            .str.title()
        )
    return df
```

```python
def clean_df(df, column_mapping, float_cols, cat_cols):
    # Step 1: Select and rename columns first
    df = df[list(column_mapping.keys())].copy()
    df = df.rename(columns=column_mapping)

    # Step 2: Clean numericals and categoricals using unified names
    df = clean_numerical_columns(df, float_cols)
    df = clean_categorical_columns(df, cat_cols)

    # Step 3: Special case cleanup
    df = clean_brand_column(df)

    return df
```

```python
cleaned_dfs = []

for year in [2012, 2013, 2014, 2015]:
    df = dfs[year]
    rename_map = rename_maps[year]
    cleaned_df = clean_df(df, rename_map, float_columns, categorical_columns)
    cleaned_dfs.append(cleaned_df)

merged_df = pd.concat(cleaned_dfs, ignore_index=True)
```

```python
merged_df = pd.concat(cleaned_dfs, ignore_index=True)
print("Merged shape:", merged_df.shape)
print(merged_df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(merged_df['co2'], bins=50, kde=True)
plt.title("Distribution of CO2 Emissions (g/km)")
plt.xlabel("CO2 (g/km)")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

#### Distribution of CO₂ Emissions

The histogram shows that the majority of vehicles emit between 180 and 220 g/km of CO₂, with a pronounced peak near 200 g/km. A secondary group of more efficient vehicles clusters between 100 and 130 g/km. Emissions above 300 g/km are rare but present, likely corresponding to performance or heavy-duty vehicles. This distribution is moderately right-skewed and suggests the need to handle outliers carefully during preprocessing and modeling.

```python
import matplotlib.pyplot as plt
import seaborn as sns


merged_df['fuel'] = merged_df['fuel'].astype(str).str.strip().str.upper()


plot_df = merged_df[['fuel', 'co2']].dropna()


fuel_order = sorted(plot_df['fuel'].unique())


plt.figure(figsize=(14, 6))
sns.boxplot(data=plot_df, x='fuel', y='co2', order=fuel_order)
plt.title("CO2 Emissions by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("CO2 Emissions (g/km)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

Observation:

Emissions vary significantly across fuel types.

Some fuels (e.g., EL, EE, EH) have very low CO2 emissions — possibly electric or hybrid types.

GO (diesel) and E5 (petrol) have higher median emissions and more spread.

Implication:

Fuel type is a strong categorical predictor.

You may want to group rare fuel types or consolidate them by family (Gasoline, Diesel, Electric, Hybrid, etc.).

```python
co2_yearly = merged_df.groupby("Year")["co2"].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.lineplot(data=co2_yearly, x='Year', y='co2', marker='o')
plt.title("Average CO2 Emissions Over Time")
plt.xlabel("Year")
plt.ylabel("Average CO2 Emissions (g/km)")
plt.grid(True)
plt.show()
```
